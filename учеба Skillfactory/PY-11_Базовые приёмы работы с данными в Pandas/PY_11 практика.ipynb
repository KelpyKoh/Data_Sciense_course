```python
mask = citi_bike_copy['age'] > 60
old_rides = citi_bike_copy[mask]
old_rides.info()
```

```python
citi_bike_copy['starttime'] = pd.to_datetime(citi_bike_copy['starttime'])
citi_bike_copy['stoptime'] = pd.to_datetime(citi_bike_copy['stoptime'])
```

```python
citi_bike_copy['trip duration'] = citi_bike_copy['stoptime'] - citi_bike_copy['starttime']
citi_bike_copy.head()
```

```python
print(citi_bike_copy['trip duration'].mean().seconds)
```

```python
citi_bike_copy['weekend'] = citi_bike_copy['starttime'].dt.dayofweek.apply(lambda x: 1 if x == 5 or x == 6 else 0)
weekend_mask = citi_bike_copy['weekend'] == 1
bike_weekend = citi_bike_copy[weekend_mask]
bike_weekend.info()
```

```python
def get_time_of_day(hours):
    if hours <= 6:
        result = 'night'
        return result
    elif hours > 6 and hours <= 12:
        result = 'morning'
        return result
    elif hours > 12 and hours <= 18:
        result = 'day'
        return result
    elif hours > 18 and hours <= 23:
        result = 'evening'
        return result


citi_bike_copy['time_of_day'] = citi_bike_copy['starttime'].dt.hour.apply(get_time_of_day)
citi_bike_copy.tail()
```

```python
count_time_of_day = citi_bike_copy['time_of_day'].value_counts()
answere = count_time_of_day['day'] / count_time_of_day['night']
print(round(answere))
```

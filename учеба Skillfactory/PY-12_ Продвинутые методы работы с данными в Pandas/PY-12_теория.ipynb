```python
import pandas as pd

melb_df = pd.read_csv('data/melb_data_fe.csv')
melb_df.head()
```

```python
display(melb_df.info())
```

```python
melb_df['Date'] = pd.to_datetime(melb_df['Date'])
```

```python
print(melb_df['Date'].dt.quarter.value_counts().nlargest(2))
#print(melb_df['Date'].dt.quarter.iloc[0:, 2]
```

```python
exclude_list = ['Date', 'Rooms', 'Bedroom', 'Bathroom', 'Car']
max_unique = 150

for col in melb_df.columns:    
    if melb_df[col].nunique() < max_unique and col not in exclude_list:
        melb_df[col] = melb_df[col].astype('category')
        

display(melb_df.info())
```

```python
display(melb_df.sort_values(by='AreaRatio', ascending=False, ignore_index=True).loc[1558, 'BuildingArea'])
```

```python
mask_type = melb_df['Type'] == 'townhouse'
mask_rooms = melb_df['Rooms'] > 2

display(melb_df[mask_type & mask_rooms].sort_values(
    by=['Rooms', 'MeanRoomsSquare'],
    ascending=[True, False],
    ignore_index=True
).loc[18, 'Price'])
```

```python
melb_df.groupby('Rooms')['Price'].mean().sort_values()
```

```python
melb_df.groupby('Regionname')['Lattitude'].std().sort_values()
```

```python
melb_df['Date'] = pd.to_datetime(melb_df['Date'])
```

```python
#summer_delta = pd.to_datetime('2017-09-01') - pd.to_datetime('2017-05-01')
summer_mask1 = melb_df['Date'].dt.date  < pd.to_datetime('2017-09-01')
summer_mask2 = melb_df['Date'].dt.date  > pd.to_datetime('2017-05-01')
melb_df_summer = melb_df[summer_mask1 & summer_mask2]

melb_df_summer.groupby('SellerG')['Price'].sum().sort_values()
```

```python
from statistics import median


melb_df.pivot_table(
    values='BuildingArea',
    index='Rooms',
    columns='Type',
    aggfunc=median
)
```
